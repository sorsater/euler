The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28.

The first ten terms would be:  

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

 1: 1  
 3: 1,3  
 6: 1,2,3,6  
10: 1,2,5,10  
15: 1,3,5,15  
21: 1,3,7,21  
28: 1,2,4,7,14,28  
We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
import math
import time
```

```python
def get_num_divisors(n):
    divisors = set()
    for i in range(1, int(math.sqrt(n)) + 1):
        if n % i == 0:
            divisors.add(i)
            divisors.add(n // i)

    return len(divisors)
```

```python
def generate_triangle_numbers(max_divisors):
    start = time.time()
    triangle_number, cur_nr = 0, 1
    while True:
        triangle_number += cur_nr
        cur_nr += 1
        if get_num_divisors(triangle_number) > max_divisors:
            break
    print(triangle_number)
    print('Took: {:.2f} seconds'.format(time.time() - start))
```

## Test

```python
generate_triangle_numbers(5)
```

## Answer

```python
generate_triangle_numbers(500)
```
