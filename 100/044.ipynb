Pentagonal numbers are generated by the formula, Pn=n(3n−1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are pentagonal and D = |Pk − Pj| is minimised; what is the value of D?

## Analysis
We loop and continously create new pentagonal numbers.  
When creating a new pentagonal (largest so far, curr_pent). We loop over all previous pentagonals (p_k).  
We calculate what the third (p_j) would be (p_k + p_j = curr_pent).  
With this new value, p_j, if this one is a pentagonal as well we have a final value to check, if p_k+p_j is a pentagonal.

Note that this solution return only the first occurence.

```python
pentagonals = set()
n = 1
found = False
while not found:
    curr_pent = int(n*(3*n-1)/2)
    pentagonals.add(curr_pent)
    for p_k in pentagonals:
        p_j = curr_pent - p_k
        if p_j in pentagonals:
            if p_k - p_j in pentagonals:
                print(p_k - p_j)
                found = True
    n += 1
```
